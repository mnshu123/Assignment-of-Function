```python
#Q4. check prime or not prime [use with argument & no return].

def prime(n):
    count=2
    while(count<n):
        if(n%count==0):
            print(n,"Is not a prime number.")
            exit()
        count +=1
        print(n,"Is prime number.")
num = int(input("enter any value:"))
prime(num) 

                     #OR
def prime(n):
    count =2
    while(2<n):
        if(n%2==0):
            print(n,"Is not prime number.")
            exit()
        count+=1
        print(n,"Is prime number.")
a = int(input("enter any value:"))
prime(a)
```

```python
#Q6. Fibonacci series :- [with function & No return value].
# Use while loop:-
def fibo(n):
    a,b=0,1
    i=2
    print(a,b,end=' ')
    while(i<n):
        c= a+b
        print(c,end=' ')
        a=b
        b=c
        i+=1
x = int(input("enter any value:"))
fibo(x)      
```

```python
#Q11.  find factorial number. [function with argument & with return value].

def fact(n):
    i =1
    while(n>0):                           # use while loop
        i = n*i
        n-=1
        return i
x = int(input("enter any value:"))
print("fact =",fact(x))
```
